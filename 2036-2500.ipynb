```python
import pandas as pd
import openai
import tqdm

openai.api_key =  ""
path = r"D:\new622_sfsk\corpus\new_622_data\score\test\test_gpt.csv"

data = pd.read_csv(path)
data = data[2040:2500]

maxlen=2800
result=[]
idxs=[]
pre_label=[]
true_label=[]
all_query=[]
true_or_false =[]

for sta,idx,q1, q2, q3, q4,true_tag in zip(tqdm.tqdm(data['statement'],leave=False), data['id'],
                                  data['option_A'], data['option_B'], data['option_C'], data['option_D'],data['answer']):
    query = "请直接给出下列问题的正确答案："+sta  + ' A.'+ q1 + ' B.'+q2 + ' C.'+q3 + ' D.'+q4
    if len(query) > maxlen:
        query=query[-maxlen:]
    rsp = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages=[
            {"role": "system", "content": "司法考试"},
            {"role": "user", "content": query} 
        ]
    )

    ans = rsp.get("choices")[0]["message"]["content"]
    print("\n"+query+"\n"+"-"*30)
    print("问题id:"+idx+"\t正确答案为："+true_tag+"\nchatGPT预测答案为："+ans)


#     result.append(ans)# 返回结果
    
    idxs.append(idx)# id
    true_label.append(true_tag)
    all_query.append(query)
    
    true_or_false.append('')

        
bm_dict = {'id':idxs,'query':all_query,'true_label':true_label, "true_or_false":true_or_false}
bm_frame = pd.DataFrame(bm_dict)


bm_frame.to_csv(r"D:\Desktop\chat_result\chatgpt2036-2500.csv", sep=',', index=False, header=True, encoding='utf_8_sig')
```
